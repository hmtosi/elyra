# Test Notebook - Directory Dependency

This notebook demonstrates using a directory as a dependency.
The `sample_data/` directory should be included as a dependency when this notebook is used in a pipeline.

```python
import os
import sys

# Add sample_data to path
sys.path.insert(0, "./sample_data")

print("Testing directory dependency...")
print(f"Current working directory: {os.getcwd()}")
```

```python
# Verify the directory exists
if os.path.exists("./sample_data"):
    print("✅ sample_data directory found!")
    print("Contents:")
    for item in os.listdir("./sample_data"):
        print(f"  - {item}")
else:
    print("❌ sample_data directory NOT found")
    print("This test requires the sample_data directory as a dependency")
```

```python
# Import and use utilities from the dependency directory
from utils import load_config, load_data, process_data

# Load configuration
config = load_config("./sample_data/config.json")
print(f"\nLoaded config: {config['model_name']} v{config['version']}")

# Load data
data = load_data("./sample_data/data.csv")
print(f"Loaded {len(data)} data rows")

# Process
result = process_data(data, config)
print(f"\n✅ Successfully processed data using directory dependency!")
```
